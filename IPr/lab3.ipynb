```python
# Использование библиотеки matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#%matplotlib inline позволяет выводить графики matplotlib в Jupyter
plt.imshow(image)
```

```python
rgb_image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
(h, w, d) = rgb_image.shape
angle = -36
scale = 1.2
center = (w // 2.5, h // 3)
M = cv.getRotationMatrix2D(center, angle, scale)
rotated = cv.warpAffine(rgb_image, M, (w, h))
plt.imshow(rotated)
```

```python
ROI = (0, 20, 152, 215)  #(x0, y0, height, width)
cropped_image = rotated[ROI[1]:ROI[1] + ROI[3], ROI[0]:ROI[0] + ROI[2]]
plt.imshow(cropped_image)
```

```python
filename = 'image_1.png'
cv.imwrite(filename, cropped_image) 
```

```python
image1 = cropped_image

plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image1) 
#Create the sharpening kernel 
kernel = np.array([[0, -1, 0], [-1, 5.3, -1], [0, -1, 0]]) 
# Sharpen the image 
sharpened_image = cv.filter2D(image1, -1, kernel) 
#Plot the sharpened image 
plt.subplot(1, 2, 2) 
plt.title("Sharpening") 
plt.imshow(sharpened_image) 
plt.show()
```
