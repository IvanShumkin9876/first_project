```python
import sys
sys.path.append(r'C:\Users\student\Desktop')
%matplotlib inline
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np

# Изменим стандартный размер графиков matplotlib
plt.rcParams["figure.figsize"] = [6, 4]
```

```python
image = cv.imread('image.png')
```

```python
# Использование библиотеки matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#%matplotlib inline позволяет выводить графики matplotlib в Jupyter
plt.imshow(image)
```

```python
image1 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image1)
```

```python
channels = [0]
histSize = [256]
range1 = [0, 256]
lut = lambda i: 1.5 * ((i - np.min(i)) / (np.max(i) - np.min(i)))
result_image = lut(image1)

gs = plt.GridSpec(2, 2)
plt.figure(figsize=(20, 16))
plt.subplot(gs[0])
plt.imshow(image1, cmap='gray')
plt.subplot(gs[1])
plt.imshow(result_image)
plt.subplot(gs[2])
plt.hist(image1.reshape(-1), 256, range1)
plt.subplot(gs[3])
plt.hist(result_image.reshape(-1), 256, range1)
plt.show()
```

```python

plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image1) 
kernel = np.array([[0, -1, 0], [-1, 9, -1], [0,-1, 0]])
# Sharpen the image 
sharpened_image = cv.filter2D(image1, -1, kernel) 
#Plot the sharpened image 
plt.figure(figsize=(20, 16))
plt.subplot(1, 2, 2) 
plt.title("Sharpening") 
plt.imshow(sharpened_image) 
plt.show()
```
